```python
### Data Transformation - Fill the SIREN
#Step 1 - find Serial Number Siren Mapping
combined_csv["legal_infos_head_office_registration_number"].dtype

combined_csv["legal_infos_head_office_registration_number"] = combined_csv["legal_infos_head_office_registration_number"].astype(str)
```

```python
SerialNumberAndSIREN =combined_csv.groupby('serial_number')['legal_infos_head_office_registration_number'].agg([('legal_infos_head_office_registration_number', '| '.join)]).reset_index()
SerialNumberAndSIREN
```

```python
def remove_dup(string):
    temp=string.split('|')
    temp=[x.strip() for x in temp]
    return '|'.join(set(temp))


def hasNaNinSIREN(string):
    if "nan" in string:
        return True
    else:
        return False

def hasOnlyNan(string):
    if "nan" in string and "|" not in string:
        return True
    else:
        return False

def hasMoreThanOneSIRENExcludingNan(string):
    temp=string.split('|')
    if "nan" in temp:
        temp.remove("nan")
    if len(temp) > 1:
        return True
    else:
        return False



SerialNumberAndSIREN["uniqueSIRENNumber"] = SerialNumberAndSIREN['legal_infos_head_office_registration_number'].apply(remove_dup)
SerialNumberAndSIREN["hasNaN"] = SerialNumberAndSIREN['legal_infos_head_office_registration_number'].apply(hasNaNinSIREN)
SerialNumberAndSIREN["hasOnlyNaN"] = SerialNumberAndSIREN['uniqueSIRENNumber'].apply(hasOnlyNan)

SerialNumberAndSIREN["hasMoreThanOneSIRENExcludingNan"] = SerialNumberAndSIREN['uniqueSIRENNumber'].apply(hasMoreThanOneSIRENExcludingNan)


```

```python
SerialNumberAndSIREN=SerialNumberAndSIREN.drop(["legal_infos_head_office_registration_number"], axis=1)
```

```python
SerialNumberAndSIREN
```

```python
def isDiscard(record):
    if(record[0]):
        return True
    if(record[1] and record[2]):
        return True
    else:
        return False

SerialNumberAndSIREN["ToDiscard"] = SerialNumberAndSIREN[["hasOnlyNaN","hasNaN","hasMoreThanOneSIRENExcludingNan"]].apply(isDiscard,axis=1)
```

```python
SerialNumberAndSIREN.to_csv( "SerialNumberAndSIREN.csv", index=False, encoding='utf-8-sig')
```
